```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("가중치데이터_최종.csv", index_col=0)
df
```

```python
df_rank = df.copy()
```

```python
df_rank['전력사용량_rank'] = df_rank['감소추이'].rank()
df_rank['지진위험도_rank'] = df_rank['지진위험도'].rank(ascending=False)
df_rank['평균기온_rank'] = df_rank['평균기온(°C)'].rank(ascending=False)
df_rank['태풍위험도_rank'] = df_rank['태풍위험도'].rank(ascending=False)
df_rank['sw기업_rank'] = df_rank['기업갯수'].rank()
df_rank['거래금액_rank'] = df_rank['거래금액(만원)'].rank(ascending=False)
```

```python
df_rank = df_rank[['시군', '전력사용량_rank', '지진위험도_rank', '평균기온_rank',
                   '태풍위험도_rank', 'sw기업_rank', '거래금액_rank']]
df_rank
```

```python
weights = [100, 10, 5, 2, 1, 0.5]
data_col = ['전력사용량_rank', '지진위험도_rank', '평균기온_rank',
            '태풍위험도_rank', 'sw기업_rank', '거래금액_rank']
```

```python
for i in range(109):
    df_rank.loc[i, "최종점수"] = sum(df_rank.loc[i, data_col] * weights)
```
