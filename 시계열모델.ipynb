### 여기부터는 블로그 참고 (아직 정확한 코드 이해X)

```python
import prophet
from prophet import Prophet
```

```python
bs = pd.DataFrame(test.iloc[:,1])
bs = bs.reset_index().rename(columns={'index':'ds','부산광역시':'y'})
bs
```

```python
bs['ds'] = bs['ds'].astype('str') + '01'
bs
```

```python
bs['ds'] = pd.to_datetime(bs['ds'], format='%Y%m%d')
bs
```

```python
model =  Prophet() 
model.fit(bs)
```

```python
past = model.make_future_dataframe(periods=0, freq='M')
future_data = pd.DataFrame(['2022-07-01','2022-08-01'], columns=['ds']).astype( 'datetime64[ns]')
future_data = pd.concat([past,future_data],axis=0)
forecast_data = model.predict(future_data)
```

```python
from prophet.plot import add_changepoints_to_plot
fig = model.plot(forecast_data)
a = add_changepoints_to_plot(fig.gca(), model, forecast_data)
```
